```python
import pandas as pd

# Load the data
file_path = '/content/Sales Data.csv'
sales_data = pd.read_csv(file_path)

# Display basic information about the dataset
print(sales_data.info())

# Display the first few rows of the dataset
print(sales_data.head())
```

```python
# Remove the 'Unnamed: 0' column
sales_data.drop(columns=['Unnamed: 0'], inplace=True)

# Check for missing values
missing_values = sales_data.isnull().sum()

# Convert 'Order Date' to datetime
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])

# Display cleaned data information
cleaned_data_info = sales_data.info()

# Display missing values information
missing_values
```

```python
import matplotlib.pyplot as plt

# Group by month and calculate total sales
# Extract the month from the 'Order Date' as a new column
sales_data['Month'] = sales_data['Order Date'].dt.month
monthly_sales = sales_data.groupby('Month')['Sales'].sum()

# Plot the monthly sales
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by city and calculate total sales
city_sales = sales_data.groupby('City')['Sales'].sum()

# Plot the city sales
plt.figure(figsize=(12, 6))
city_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by product and calculate total quantity ordered
top_products_quantity = sales_data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)

# Plot the top products by quantity sold
plt.figure(figsize=(12, 6))
top_products_quantity.plot(kind='bar', color='coral')
plt.title('Top Products by Quantity Sold')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by product and calculate total sales
top_products_sales = sales_data.groupby('Product')['Sales'].sum().sort_values(ascending=False)

# Plot the top products by total sales
plt.figure(figsize=(12, 6))
top_products_sales.plot(kind='bar', color='green')
plt.title('Top Products by Total Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html

# Load the data
file_path = '/content/Sales Data.csv'
sales_data = pd.read_csv(file_path)

# Convert 'Order Date' to datetime
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'], errors='coerce')

# Extract additional time-related features
sales_data['Day'] = sales_data['Order Date'].dt.day
sales_data['Month'] = sales_data['Order Date'].dt.month
sales_data['Hour'] = sales_data['Order Date'].dt.hour

# Total sales by month
monthly_sales = sales_data.groupby('Month')['Sales'].sum().reset_index()

# Total sales by day
daily_sales = sales_data.groupby('Day')['Sales'].sum().reset_index()

# Total sales by hour
hourly_sales = sales_data.groupby('Hour')['Sales'].sum().reset_index()

# Top products by quantity sold
top_products_quantity = sales_data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).reset_index()

# Top products by total sales
top_products_sales = sales_data.groupby('Product')['Sales'].sum().sort_values(ascending=False).reset_index()

# Initialize the Dash app
app = Dash(__name__)

app.layout = html.Div([
    html.H1('Sales Data Analysis Dashboard'),

    html.Div([
        html.H2('Monthly Sales'),
        dcc.Graph(
            id='monthly-sales',
            figure=px.bar(monthly_sales, x='Month', y='Sales', title='Total Sales by Month', labels={'Sales': 'Sales', 'Month': 'Month'})
        )
    ]),

    html.Div([
        html.H2('Daily Sales'),
        dcc.Graph(
            id='daily-sales',
            figure=px.line(daily_sales, x='Day', y='Sales', title='Total Sales by Day', labels={'Sales': 'Sales', 'Day': 'Day'})
        )
    ]),

    html.Div([
        html.H2('Hourly Sales'),
        dcc.Graph(
            id='hourly-sales',
            figure=px.bar(hourly_sales, x='Hour', y='Sales', title='Total Sales by Hour', labels={'Sales': 'Sales', 'Hour': 'Hour'})
        )
    ]),

    html.Div([
        html.H2('Hourly Sales Distribution'),
        dcc.Graph(
            id='hourly-sales-pie',
            figure=px.pie(hourly_sales, names='Hour', values='Sales', title='Sales Distribution by Hour')
        )
    ]),

    html.Div([
        html.H2('Top Products by Quantity Sold'),
        dcc.Graph(
            id='top-products-quantity',
            figure=px.bar(top_products_quantity, x='Product', y='Quantity Ordered', title='Top Products by Quantity Sold', labels={'Quantity Ordered': 'Quantity Ordered', 'Product': 'Product'})
        )
    ]),

    html.Div([
        html.H2('Top Products by Total Sales'),
        dcc.Graph(
            id='top-products-sales',
            figure=px.bar(top_products_sales, x='Product', y='Sales', title='Top Products by Total Sales', labels={'Sales': 'Sales', 'Product': 'Product'})
        )
    ])
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
